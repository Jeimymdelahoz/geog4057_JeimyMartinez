```python
aprx = arcpy.mp.ArcGISProject("CURRENT")
m = aprx.listMaps("Downtown")[0]
lyr = m.listLayers("parks")[0]

sym = lyr.symbology

green = {"RGB": [100, 175, 0, 100]}  # starting color

if lyr.isFeatureLayer and hasattr(sym, "renderer"):
    sym.renderer.symbol.color = green
    lyr.symbology = sym

del aprx
```

What type is the 'green' variable? What do the numbers in the list mean? If you want to show a brown color, what numbers you would fill it with? Change it in the block and confirm the color is changed to brown.

The green variable is a Python dictionary.
The numbers inside the list ([100, 175, 0, 100]) represent an RGBA color, meaning:

R = 100 (red)

G = 175 (green)

B = 0 (blue)

A = 100 (alpha/transparency)

```python
# Create a Layout and Insert the Downtown Map
aprx = arcpy.mp.ArcGISProject("CURRENT")
m = aprx.listMaps("Downtown")[0]

lyt = aprx.createLayout(11, 8.5, 'INCH', 'New Layout with Rectangles')

def MakeRec_LL(llx, lly, w, h):
    xyRecList = [[llx, lly], [llx, lly+h], [llx+w,lly+h], [llx+w,lly], [llx,lly]]
    array = arcpy.Array([arcpy.Point(*coords) for coords in xyRecList])
    rec = arcpy.Polygon(array)
    return rec

mf = lyt.createMapFrame(MakeRec_LL(0.5, 0.5, 10, 7.5), m, "New Map Frame")

del aprx
```

1. What is the size of the layout in inches?
11 inches × 8.5 inches

2. What is the size of the map frame?
10 inches × 7.5 inches

3. Comparing manual vs Python-created layouts: what are the advantages of creating it in Python?

The layout created in Python is sized exactly at 11×8.5 inches, and the map frame is exactly 10×7.5 inches. Compared to creating layouts manually in ArcGIS Pro, using Python has several advantages:

Precision: Python allows exact control over layout and map frame dimensions.

Reproducibility: The same code can recreate the layout at any time, guaranteeing consistent results.

Automation: Python can generate multiple layouts quickly without manually configuring each one.

Efficiency: Complex or repetitive layout tasks can be done in seconds through scripting.

Overall, creating the layout with Python is more accurate and scalable than manual creation.


```python
aprx = arcpy.mp.ArcGISProject("CURRENT")


lyt = aprx.listLayouts("New Layout with Rectangles")[0]
m = aprx.listMaps("Downtown")[0]

# Recreate MakeRec_LL if needed
def MakeRec_LL(llx, lly, w, h):
    xyRecList = [[llx, lly], [llx, lly+h], [llx+w, lly+h], [llx+w, lly], [llx, lly]]
    array = arcpy.Array([arcpy.Point(*coords) for coords in xyRecList])
    rec = arcpy.Polygon(array)
    return rec

# Get the map frame inside the layout
mf = lyt.listElements("MAPFRAME_ELEMENT", "New Map Frame")[0]

# ------------------------------
# Create a North Arrow
# ------------------------------
naStyle = aprx.listStyleItems('ArcGIS 2D', 'North_Arrow', 'Compass North 1')[0]
na = lyt.createMapSurroundElement(
    arcpy.Point(9.5, 7.5),
    'North_Arrow',
    mf,
    naStyle,
    "Compass North Arrow"
)
na.elementWidth = 0.5

# ------------------------------
# Create a Scale Bar
# ------------------------------
sbName = 'Double Alternating Scale Bar 1 Metric'
sbStyle = aprx.listStyleItems('ArcGIS 2D', 'Scale_bar', sbName)[0]
sbEnv = MakeRec_LL(5.5, 0.1, 4, 0.5)

sb = lyt.createMapSurroundElement(
    sbEnv,
    'Scale_bar',
    mf,
    sbStyle,
    'New Scale Bar'
)

del aprx
```

```python
#  create a legend
aprx = arcpy.mp.ArcGISProject("CURRENT")

# Get the existing layout
lyt = aprx.listLayouts("New Layout with Rectangles")[0]

# Get the map frame
mf = lyt.listElements("MAPFRAME_ELEMENT", "New Map Frame")[0]

# ------------------------------
# Create a Legend
# ------------------------------
legSi = aprx.listStyleItems('ArcGIS 2D', 'LEGEND', 'Legend 3')[0]

leg = lyt.createMapSurroundElement(
    arcpy.Point(1, 7),  # position
    'LEGEND',
    mf,
    legSi,
    'New Legend Element'
)

# Modify legend properties
leg.elementWidth = 3
leg.elementHeight = 3
leg.fittingStrategy = 'AdjustFontSize'
leg.columnCount = 1
leg.title = 'Downtown'

del aprx
```
